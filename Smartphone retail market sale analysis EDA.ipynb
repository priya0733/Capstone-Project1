```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/SMARTPHONE RETAIL OUTLET SALE DATA.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
#duplicate data
data.duplicated()
```

```python
#plot scatter
plt.scatter(data['Price'], data['Amount'])
plt.title('Scatter Plot')
plt.xlabel('Price')
plt.ylabel('Amount')
plt.grid()
plt.show()
```

```python
#histogram
plt.hist(data['Price'])
plt.title('Histogram')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#piechart
product_count = data['TYPE OF PRODUCT'].value_counts()
plt.pie(product_count, labels=product_count.index, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
product_sales = data.groupby('TYPE OF PRODUCT')['Amount'].sum()
plt.bar(product_sales.index, product_sales.values)
plt.title('Product Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
plt.hist(data['Amount'], bins=5)
plt.title('Distribution of Sales Amounts')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
payment_method_sales = data.groupby('PAYMENT TYPE')['Amount'].sum()
plt.pie(payment_method_sales.values, labels=payment_method_sales.index, autopct='%1.1f%%')
plt.title('Sales by Payment Method')
plt.show()
```

```python
# Histograms for numerical features
sns.histplot(data['Price'])
plt.show()
sns.histplot(data['Amount'])
plt.show()
```

```python
# Box plots for numerical features by categorical features
sns.boxplot(x='TYPE OF PRODUCT', y='Amount', data=data)
plt.show()
```

```python
# Bar plots for categorical features
sns.countplot(x='PAYMENT TYPE', data=data)
plt.show()
```

```python
sns.boxplot(x=data['Price'])
plt.show()
```
